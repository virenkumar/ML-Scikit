Since there are several values missing for GarageType and garage related features and BasementQual and other basement related columns, we'll create another category Unknown for these features. We can later compare it with another strategy, if needed

```python
def create_missing_class(housing_data):
    #housing_n = housing_data.fillna({'GarageType':'unknown', 'GarageFinish':'unknown', 'GarageQual':'unknown', 'GarageCond':'unknown', 'BsmtQual':'unknown', 'BsmtCond':'unknown', 'BsmtExposure':'unknown', 'BsmtFinType1':'unknown', 'BsmtFinType2':'unknown' })
    housing_n = housing_data.fillna('unknown')
    return housing_n
```

```python
housing_catmod = create_missing_class(housing_cat)
housing_catmod.head()
```

```python
housing_catmod.isnull().sum()
```

Now only 8 values for MasVnrType and 1 for Electrical are NaN. We can remove these rows since there will be still sufficient data available 

```python
def drop_missingrows(housing_data):
    housing_n = housing_data.dropna(subset=['MasVnrType', 'Electrical'])
    return housing_n;
```

```python
housing_catmod = drop_missingrows(housing_catmod)
housing_catmod.head()
```

```python
housing_catmod.isnull().sum()
```

Now the data is clean, we can apply other transformations

```python
try:
    from sklearn.preprocessing import OrdinalEncoder # just to raise an ImportError if Scikit-Learn < 0.20
    from sklearn.preprocessing import OneHotEncoder
except ImportError:
    from future_encoders import OneHotEncoder # Scikit-Learn < 0.20

cat_encoder = OneHotEncoder(handle_unknown='ignore')
housing_cat1hot = cat_encoder.fit_transform(housing_catmod)
housing_cat1hot
```

```python
housing_cat1hot.toarray()
```

Build the Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])
```

```python
housing_num_tr = num_pipeline.fit_transform(housing_num)
housing_num_tr
```

```python
from sklearn.preprocessing import FunctionTransformer
create_missing_class_transformer = FunctionTransformer(create_missing_class)
```

```python

cat_pipeline = Pipeline([
        ('unknown_cat', create_missing_class_transformer),
        ('encode', OneHotEncoder(handle_unknown='ignore')),
    ])
```

```python
try:
    from sklearn.compose import ColumnTransformer
except ImportError:
    from future_encoders import ColumnTransformer 
```

```python
# There should be an IMPUTER for Categorical also - since we've used the strategy of creating Unknown class etc. ??????????
num_attribs = list(housing_num)
cat_attribs = list(housing_cat)

full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", cat_pipeline, cat_attribs),
    ])
num_attribs
```

```python
#remove the rows which need not be considered
train_datamod = drop_missingrows(train_data)
```

```python
y_train  = train_datamod["SalePrice"].copy()
X_train = train_datamod.drop("SalePrice", axis=1)
```

```python
housing_prepared = full_pipeline.fit_transform(X_train)
housing_prepared.shape
```
